```python
import pandas as pd

percorso_csv = 'C:/Users/longo/Downloads/owid-covid-data.csv'

db_covid = pd.read_csv(percorso_csv, low_memory=False)

db_covid
```

```python
# Si richiede di verificare le dimensioni del dataset e le diciture presenti nell'intestazione;

dimensioni = db_covid.shape
numero_righe, numero_colonne = dimensioni

print(f"Numero di righe: {numero_righe}")
print(f"Numero di colonne: {numero_colonne}")

diciture = db_covid.columns.tolist()
print("Diciture presenti nell'intestazione:")
for dicitura in diciture:
    print(dicitura)
```

```python
# Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente;

db_covid = db_covid.dropna(subset=['continent'])

casi_totali_per_continente = db_covid.groupby('continent')['total_cases'].sum()

print("Numero totale di casi per ogni continente:")
print(casi_totali_per_continente)
```

```python
# si chiede di sviluppare una funzione che prenda in input il dataset e due nomi di continenti, e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, media, e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente);

def confronta_descrittori_statistici(db_covid, continente1, continente2):
    subset_dataset = db_covid[db_covid['continent'].isin([continente1, continente2])]
    
    descrittori_statistici = subset_dataset['total_cases'].describe()

    totale_casi_mondo = db_covid['total_cases'].sum()
    percentuale_rispetto_al_mondo = (descrittori_statistici / totale_casi_mondo) * 100
    percentuale_rispetto_al_mondo = percentuale_rispetto_al_mondo.apply(lambda x: f'{x:.3f}%')


    risultati = pd.DataFrame({
        f'{continente1}_vs_{continente2}': descrittori_statistici,
        '      Percentuale_rispetto_al_mondo': percentuale_rispetto_al_mondo
    })

    return risultati

risultati_confronto = confronta_descrittori_statistici(db_covid, 'Europe', 'South America')

print("Descrittori statistici e percentuali di casi totali:")
print(risultati_confronto)
```

```python
def confronta_vaccinazioni_tutti(db_covid):
    # lista completa dei continenti nel dataset
    lista_continenti = db_covid['continent'].unique()

    # Iniziazzazione dizionario
    risultati_vaccinazioni = {}

    # Iterare su tutte le coppie di continenti possibili
    for i in range(len(lista_continenti)):
        for j in range(i + 1, len(lista_continenti)):
            continente1 = lista_continenti[i]
            continente2 = lista_continenti[j]

            # Filtroil dataset per i due continenti specificati
            subset_dataset = db_covid[db_covid['continent'].isin([continente1, continente2])]

            # Calcolo i descrittori statistici per le vaccinazioni totali
            descrittori_statistici_vaccinazioni = subset_dataset['total_vaccinations'].describe()

            # Calcolo la percentuale rispetto al numero totale di vaccinazioni nel mondo
            totale_vaccinazioni_mondo = db_covid['total_vaccinations'].sum()
            percentuale_rispetto_al_mondo_vaccinazioni = (descrittori_statistici_vaccinazioni / totale_vaccinazioni_mondo) * 100

            # Aggiungo risultato al dizionario
            chiave_risultato = f'{continente1}_vs_{continente2}_Vaccinazioni'
            risultati_vaccinazioni[chiave_risultato] = percentuale_rispetto_al_mondo_vaccinazioni

    # Creo un DataFrame con i risultati
    risultati_df = pd.DataFrame(risultati_vaccinazioni)

    return risultati_df

# Esempio di utilizzo della funzione
risultati_vaccinazioni_tutti_confronto = confronta_vaccinazioni_tutti(db_covid)

# Stampa i risultati
print("Descrittori statistici e percentuali di vaccinazioni totali:")
print(risultati_vaccinazioni_tutti_confronto)
```
